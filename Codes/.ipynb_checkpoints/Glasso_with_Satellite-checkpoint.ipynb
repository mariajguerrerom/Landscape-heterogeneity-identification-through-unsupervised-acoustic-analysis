organizo dataframes diferentes para cada indice, estadística y buffer

```python
# Directorio donde se encuentran los archivos CSV
directory = '/Users/mariajoseguerrero/Documents/Universidad/PhD/Analisis KALE/Indices Satelitales/Estadisticas indices'
# listar todos los archivos CSV en el directorio
csv_files = [f for f in os.listdir(directory) if f.endswith('.csv')]

# Función para procesar un archivo CSV usando la coma (',') como delimitador
def procesar_csv(file_path):
    df = pd.read_csv(file_path, delimiter=',')
    # Extrayendo el tamaño del buffer y el nombre del índice del nombre del archivo
    file_name = os.path.basename(file_path)
    match = re.search(r'(\d{2,3})m_(\w+)\.csv', file_name)
    if match:
        buffer_size = match.group(1)
        index_name = match.group(2)
    else:
        raise ValueError("El nombre del archivo no sigue el patrón esperado")

    # Añadiendo columnas para el índice y el tamaño del buffer
    df['Índice'] = index_name
    df['Buffer'] = buffer_size
    return df

# Procesando todos los archivos CSV en el directorio
dataframes = []
for file in csv_files:
    file_path = os.path.join(directory, file)
    df = procesar_csv(file_path)
    dataframes.append(df)

# Concatenando todos los dataframes en uno solo
df_total = pd.concat(dataframes)

# Creando los dataframes para cada combinación de estadística, buffer e índice
lista_dataframes = []
for indice in df_total['Índice'].unique():
    for estadistica in estadisticas:
        for buffer in buffers:
            df_filtrado = df_total[(df_total['Índice'] == indice) & (df_total['Buffer'] == buffer)]
            if not df_filtrado.empty:
                df_pivot = df_filtrado.pivot_table(index=['Point', 'latitude', 'longitude'], values=estadistica, aggfunc='first')
                lista_dataframes.append({
                    "indice": indice,
                    "estadistica": estadistica,
                    "buffer": buffer,
                    "dataframe": df_pivot
                })

# Mostrando un ejemplo: el primer dataframe de la lista con su metadata
ejemplo_lista= lista_dataframes[100] if lista_dataframes else "No hay dataframes disponibles."
ejemplo_lista
```

Ejecutar solo para guardar los archvivos

```python
# Función para guardar cada dataframe en un archivo de Excel
ruta_out= '/Users/mariajoseguerrero/Documents/Universidad/PhD/Analisis KALE/Indices Satelitales/Estadisticas organizadas'
def guardar_dataframes_excel_ajustada(lista_dataframes, directorio_salida):
    rutas_archivos_guardados = []
    for df_info in lista_dataframes:
        try:
            # Verificando que el dataframe no esté vacío antes de guardarlo
            if not df_info['dataframe'].empty:
                nombre_archivo = f"{df_info['indice']}_{df_info['estadistica']}_{df_info['buffer']}m.xlsx"
                ruta_completa = os.path.join(directorio_salida, nombre_archivo)
                df_info['dataframe'].to_excel(ruta_completa)
                rutas_archivos_guardados.append(ruta_completa)
        except Exception as e:
            print(f"Error al guardar el dataframe {df_info['indice']}_{df_info['estadistica']}_{df_info['buffer']}m: {e}")
    return rutas_archivos_guardados

# Guardando cada dataframe en un archivo de Excel con el ajuste
rutas_guardadas_ajustadas = guardar_dataframes_excel_ajustada(lista_dataframes, ruta_out)

# Mostrando las rutas de los archivos guardados como ejemplo
rutas_guardadas_ajustadas[:5]  # Mostrando solo las primeras 5 rutas como ejemplo
```

```python
# Función ajustada para guardar los dataframes en archivos de Excel según la estadística
# Cada archivo de Excel tendrá hojas diferentes para cada buffer
# Las columnas serán 'Point', 'Latitude', 'Longitude', seguidas por el valor de cada índice
ruta_out='/Users/mariajoseguerrero/Documents/Universidad/PhD/Analisis KALE/Indices Satelitales/Estadisticas indices/datos organizados'

def guardar_por_estadistica_en_excel_ajustada(lista_dataframes, directorio_salida):
    rutas_archivos_guardados = []
    estadisticas_unicas = set(df_info['estadistica'] for df_info in lista_dataframes)

    for estadistica in estadisticas_unicas:
        # Creando un archivo de Excel para cada estadística
        nombre_archivo_excel = f"{estadistica}.xlsx"
        ruta_archivo_excel = os.path.join(directorio_salida, nombre_archivo_excel)
        writer = pd.ExcelWriter(ruta_archivo_excel, engine='xlsxwriter')

        for buffer in buffers:
            # Creando una hoja para cada buffer
            hoja_nombre = f"{buffer}m"
            # Inicializando un dataframe vacío para la hoja
            df_hoja = pd.DataFrame()

            for df_info in lista_dataframes:
                if df_info['estadistica'] == estadistica and df_info['buffer'] == buffer:
                    # Añadiendo los datos de cada índice al dataframe de la hoja
                    df_temp = df_info['dataframe'].reset_index()
                    df_temp.rename(columns={estadistica: df_info['indice']}, inplace=True)
                    df_hoja = pd.merge(df_hoja, df_temp, on=['Point', 'latitude', 'longitude'], how='outer') if not df_hoja.empty else df_temp

            if not df_hoja.empty:
                df_hoja.to_excel(writer, sheet_name=hoja_nombre, index=False)

        writer.save()
        rutas_archivos_guardados.append(ruta_archivo_excel)

    return rutas_archivos_guardados

# Guardando los dataframes en archivos de Excel según la nueva especificación ajustada
rutas_guardadas_por_estadistica_ajustadas = guardar_por_estadistica_en_excel_ajustada(lista_dataframes, ruta_out)

# Mostrando las rutas de los archivos guardados como ejemplo
rutas_guardadas_por_estadistica_ajustadas
```

Genero grafo para cada estadistica y cada buffer

```python
#función para plotear puntos conservando ubicación geografica
def same_posicion_data(df):
  gps_point = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude)) #objeto geometrico para respetar posiciones
  coordinates = np.column_stack((gps_point.geometry.x, gps_point.geometry.y))
  return coordinates
```

Cargo dataframe mean para 50m, 100m y 200m

```python
scaler = MinMaxScaler()
file_path = '/Users/mariajoseguerrero/Documents/Universidad/PhD/Analisis KALE/Indices Satelitales/Estadisticas indices/datos organizados/mean.xlsx'

# Lista de nombres de las hojas de cálculo
sheets = ['50m', '100m', '200m']

# Lista para almacenar las matrices de adyacencia
adjacency_matrices = []

for sheet in sheets:
    # Cargar los datos
    df = pd.read_excel(file_path, sheet_name=sheet)

    # Separar y normalizar los datos
    df2 = df.iloc[:,4:]
    df3 = df.iloc[:,0:4]
    df_norm = pd.DataFrame(scaler.fit_transform(df2), columns=df2.columns)

    # Procesar los datos
    datos = np.array(df_norm)
    datos2 = np.transpose(datos)
    datos2 /= datos2.std(axis=0)
    edge_model = GraphicalLassoCV(cv=10, max_iter=1000, verbose=True, assume_centered=True)
    edge_model.fit(datos2)
    
    print(edge_model.alpha_)

    # Crear la matriz de precisión y la visualización
    p_satelite = np.around(edge_model.precision_, decimals=3)
    p_satelite = pd.DataFrame(p_satelite)
    print(p_satelite)
    A= np.around(edge_model.covariance_, decimals=3) #covarianza
    A=pd.DataFrame(A)

    plt.figure(figsize=(15,10))
    plt.imshow(np.abs(p_satelite), cmap='coolwarm')
    plt.colorbar()
    plt.title(f"Precision Matrix for {sheet}")
    plt.show()

    # Crear el grafo
    links = p_satelite.stack().reset_index()
    links.columns = ['var1', 'var2','value']
    links2 = p_satelite.stack().reset_index()
    links2.columns = ['var1', 'var2','value']

    threshold = 0.7
    links2 = links2.loc[(links2['var1'] != links2['var2'])]
    links = links.loc[(abs(links['value']) >= threshold) & (links['var1'] != links['var2'])]

    # Crear el grafo a partir de los enlaces
    G = nx.from_pandas_edgelist(links, 'var1', 'var2', edge_attr='value', create_using=nx.Graph())

    # Obtener todos los nodos en el DataFrame de enlaces
    all_nodes = pd.Series(pd.concat([links2['var1'], links2['var2']])).unique()

    # create complete graph with all nodes
    G_complete = nx.complete_graph(all_nodes)

    # create subgraph with only significant nodes
    G_sub = G_complete.subgraph(G.nodes())

    #posición de los nodos y colores
    coordinates=same_posicion_data(df3)
    coordinates=coordinates[G_complete.nodes]
    colores=df3.Habitat.values[G_complete.nodes]
    positions = dict(zip(G_complete.nodes, coordinates))

    # Crear una lista de los pesos de los enlaces
    edge_weights = [abs(G[u][v]['value']) for u, v in G.edges()]

    plt.figure(figsize=(15,10))

    # Dibujar los nodos del grafo con tamaño personalizado
    node_size = 1400
    node_color = colores
    node_border_color = 'white'
    node_border_width = 0
    node_labels = {node: node for node in G_complete.nodes()}
    nx.draw_networkx_nodes(G_complete, positions, nodelist=G_complete.nodes(), node_size=node_size, node_color=node_color,
                           linewidths=node_border_width, edgecolors=node_border_color, cmap="summer_r")
    nx.draw_networkx_labels(G_complete, positions, labels=node_labels, font_color='black')


    # Definir la escala de colores para los enlaces
    cmap = mcolors.LinearSegmentedColormap.from_list("", ["blue", "red"]) # si valor de enlace es cercano al umbral, el color se acercará al azul, si es mucho mayor al umbral se acerca al rojo

    # Obtener una lista de colores para los enlaces
    edge_colors = [cmap(w) for w in edge_weights]

    # escalar los anchos de los enlaces
    min_width = 1  # ancho mínimo de enlace
    max_width = 5  # ancho máximo de enlace
    scaled_widths = [(w - min(edge_weights)) / (max(edge_weights) - min(edge_weights)) * (max_width - min_width) + min_width for w in edge_weights] #si valor de enlace es cercano al umbral, el color se acercará al azul, si es mucho mayor al umbral se acerca al rojo


    # Dibujar los enlaces del grafo con ancho constante y colores personalizados
    nx.draw_networkx_edges(G, positions, width=scaled_widths, edge_color="darkblue", alpha=0.8) # se puede ver tambien con colores cambiando edge_color=edge_colors
    plt.savefig(f'Glasso_indices_satelite{sheet}_mean.svg', format='svg') # cambiar nombre
    
    # Crear matriz de adyacencia vacía
    adj_matrix = np.zeros((len(G_complete), len(G_complete)))

    # Llenar la matriz de adyacencia
    for u, v in G.edges():
        adj_matrix[u, v] = 1
        adj_matrix[v, u] = 1

    # Añadir la matriz de adyacencia a la lista
    adjacency_matrices.append(adj_matrix)
```

```python
# Filtrar los enlaces según un umbral determinado
threshold = 0.3
links2 = links2.loc[(links2['var1'] != links2['var2'])]
links = links.loc[(abs(links['value']) >= threshold) & (links['var1'] != links['var2'])]

# Crear el grafo a partir de los enlaces
G = nx.from_pandas_edgelist(links, 'var1', 'var2', edge_attr='value', create_using=nx.Graph())

# Obtener todos los nodos en el DataFrame de enlaces
all_nodes = pd.Series(pd.concat([links2['var1'], links2['var2']])).unique()

# create complete graph with all nodes
G_complete = nx.complete_graph(all_nodes)

# create subgraph with only significant nodes
G_sub = G_complete.subgraph(G.nodes())

#posición de los nodos y colores
coordinates=same_posicion_data(df_info)
coordinates=coordinates[G_complete.nodes]
colores=df_info.Habitat.values[G_complete.nodes]
positions = dict(zip(G_complete.nodes, coordinates))

# Crear una lista de los pesos de los enlaces
edge_weights = [abs(G[u][v]['value']) for u, v in G.edges()]

plt.figure(figsize=(15,10))

# Dibujar los nodos del grafo con tamaño personalizado
node_size = 1400
node_color = colores
node_border_color = 'white'
node_border_width = 0
node_labels = {node: node for node in G_complete.nodes()}
nx.draw_networkx_nodes(G_complete, positions, nodelist=G_complete.nodes(), node_size=node_size, node_color=node_color,
                       linewidths=node_border_width, edgecolors=node_border_color, cmap="summer_r")
nx.draw_networkx_labels(G_complete, positions, labels=node_labels, font_color='black')


# Definir la escala de colores para los enlaces
cmap = mcolors.LinearSegmentedColormap.from_list("", ["blue", "red"]) # si valor de enlace es cercano al umbral, el color se acercará al azul, si es mucho mayor al umbral se acerca al rojo

# Obtener una lista de colores para los enlaces
edge_colors = [cmap(w) for w in edge_weights]

# escalar los anchos de los enlaces
min_width = 1  # ancho mínimo de enlace
max_width = 5  # ancho máximo de enlace
scaled_widths = [(w - min(edge_weights)) / (max(edge_weights) - min(edge_weights)) * (max_width - min_width) + min_width for w in edge_weights] #si valor de enlace es cercano al umbral, el color se acercará al azul, si es mucho mayor al umbral se acerca al rojo


# Dibujar los enlaces del grafo con ancho constante y colores personalizados
nx.draw_networkx_edges(G, positions, width=scaled_widths, edge_color="darkblue", alpha=0.8) # se puede ver tambien con colores cambiando edge_color=edge_colors
```

```python
# Crear matriz de adyacencia vacía de tamaño 11x11
adj_matrix_Glasso_sonotipos = np.zeros((len(G_complete), len(G_complete)))

enlaces = G.edges

# Llenar la matriz de adyacencia en las posiciones correspondientes a los enlaces
for enlace in enlaces:
    fila, columna = enlace
    adj_matrix_Glasso_sonotipos[fila, columna] = 1
    adj_matrix_Glasso_sonotipos[columna, fila] = 1

adj_matrix_Glasso_sonotipos_=pd.DataFrame(adj_matrix_Glasso_sonotipos)
adj_matrix_Glasso_sonotipos_
```
